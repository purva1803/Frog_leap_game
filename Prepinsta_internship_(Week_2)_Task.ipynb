```python
positions = ['G', 'G', 'G', '-', 'B', 'B', 'B']
print("Current status of the game:")
print("[ 0    1    2    3    4    5    6]")
print(positions)

empty_leaf_index = positions.index('-')

while True:
    pos = input("Press 'q' to quit, else enter the position of the piece:")

    if pos == 'q':
        print("You Quit")
        break

    pos = int(pos)

    if not 0 <= pos <= 6:
        print("Invalid Move")
        continue

    selected_position = pos

    if positions[selected_position] == 'G':

        # Conditions for 'G' frog
        if selected_position + 1 <= 6 and positions[selected_position + 1] == '-':
            pos2 = selected_position + 1
        elif selected_position + 2 <= 6 and positions[selected_position + 2] == '-' and positions[selected_position + 1] == 'B':
            pos2 = selected_position + 2
        else:
            print("Invalid Move")
            continue
    elif positions[selected_position] == 'B':

        # Conditions for 'B' frog
        if selected_position - 1 >= 0 and positions[selected_position - 1] == '-':
            pos2 = selected_position - 1
        elif selected_position - 2 >= 0 and positions[selected_position - 2] == '-' and positions[selected_position - 1] == 'G':
            pos2 = selected_position - 2
        else:
            print("Invalid Move")
            continue
    else:
        print("Invalid Frog Selection")
        continue

    # Swap the elements at selected_position and pos2
    positions[selected_position], positions[pos2] = positions[pos2], positions[selected_position]
    print("Updated status of the game:")
    print("[ 0    1    2    3    4    5    6]")
    print(positions)

    #  winning condition
    if positions == ['B', 'B', 'B', '-', 'G', 'G', 'G']:
        print("You Win!")
        break
```
